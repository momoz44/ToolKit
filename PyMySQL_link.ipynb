```python
try:
    with connection.cursor() as cursor:
        sql = 'Select com_id, trade_type from obj_sales_company_title'
        cursor.execute(sql)
        result = cursor.fetchall()
        print result
    connection.commit()
    
finally:
    connection.close();
```
